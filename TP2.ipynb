```python
import numpy as np
import matplotlib.pyplot as plt 
import os
```

```python
images_name = os.listdir(os.path.join(os.getcwd(), "images"))
```

```python
imgages = {image_name[:image_name.index(".")] : plt.imread(os.path.join(os.getcwd(), "images", image_name)) for image_name in images_name}

hist_bins = { image[0]: np.histogram(image[1].ravel(),256,[0,255]) for image in imgages.items()}
```

```python
for imgage in imgages:
    plt.hist(imgages[imgage].ravel(),256,[0,256]);
    plt.title(imgage)
    plt.show()
```
