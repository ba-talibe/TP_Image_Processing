```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft

duree_signal=100
duree_avec_padding=1000
signal=np.sin(2*np.pi*0.1*np.arange(100))
signal_padde=np.concatenate((signal,np.zeros(duree_avec_padding-duree_signal)))
plt.figure(1)
# crearion de point et norlisation
f=np.arange(duree_avec_padding)/duree_avec_padding
plt.plot(f,np.abs(fft(signal_padde)))
plt.figure(2)
plt.plot(f-1/2,np.abs(np.fft.fftshift(fft(signal_padde))))
plt.xlim([-0.2 , 0.2])
# NB : lajout de zeros  un signal, pour augmenter la rsolution frquentielle
# de la TFD, peut tre simul sns explicitement ajouter des 0
plt.figure(3)
plt.plot(f-1/2,abs(np.fft.fftshift(fft(signal,duree_avec_padding))))
plt.xlim([-0.2 , 0.2])
```

```python
from scipy.signal import hanning, hamming, blackman
```

## superposition les signaux temporels 

```python
K = 100

f = np.linspace(-np.pi, np.pi, 100)
signal_hamming = hamming(K)
signal_haning = hanning(K)
signal_blackman = blackman(K)
plt.figure(1, figsize=(10, 8))
plt.plot(f, signal_hamming, label="signal de Hamming")
plt.plot(f, signal_haning, label="signal de Hanning")
plt.plot(f, signal_blackman, label="signal de Blackman")
plt.plot(f, np.ones(K), label="signal rectangle")
plt.legend()
plt.show()
```

## Rsultat de la multiplication dun sinus par chacune des fentres

```python
K = 100
f = np.linspace(-np.pi, np.pi, K)*10


plt.figure(1, figsize=(18, 9))
plt.plot(f, np.sin(f)*hamming(K), label="signal de produit  sinus*Hamming")
plt.plot(f, np.sin(f)*hanning(K), label="signal de produit  sinus*Hanning")
plt.plot(f, np.sin(f)*blackman(K), label="signal de de produit  sinus*Blackman")
plt.plot(f, np.sin(f), label="sinus")
plt.plot(f, np.ones(K), label="signal rectangle")
plt.legend()
plt.show()
```

## spectres de chacune des fentres  une echelle lineaire

```python
K = 100

f = np.linspace(-np.pi, np.pi, 1000)
signal_hamming = hamming(K)
signal_haning = hanning(K)
signal_blackman = blackman(K)


plt.figure(figsize=(18, 8))
plt.plot(f,np.abs(np.fft.fftshift(fft(signal_hamming, 1000))), label="spectre de hamming")
plt.plot(f,np.abs(np.fft.fftshift(fft(signal_haning, 1000))),  label="spectre de hanning")
plt.plot(f,np.abs(np.fft.fftshift(fft(signal_blackman, 1000))),  label="spectre de blackman")
plt.xlim([-.5, .5])
plt.legend()
plt.show()
```

## spectres de chacune des fentres logarithmique

```python
K = 100

f = np.arange(1000)
signal_hamming = hamming(K)
signal_haning = hanning(K)
signal_blackman = blackman(K)

puissance_hanning = np.sum(signal_haning**2)/signal_haning.size 
puissance_hamming = np.sum(signal_haning**2)/signal_hamming.size 
puissance_blackman = np.sum(signal_haning**2)/signal_blackman.size 


plt.figure(figsize=(18, 8))
plt.plot(f,20*np.log(np.abs(np.fft.fftshift(fft(signal_haning, 1000)))),  label="spectre de hanning")
plt.plot(f,20*np.log(np.abs(np.fft.fftshift(fft(signal_hamming, 1000)))), label="spectre de hamming")
plt.plot(f,20*np.log(np.abs(np.fft.fftshift(fft(signal_blackman, 1000)))),  label="spectre de blackman")


plt.legend()
plt.show()
```

```python
K = 100

f = (np.arange(10000)/10000)-0.5
signal_hamming = hamming(K)
signal_haning = hanning(K)
signal_blackman = blackman(K)
rect = np.ones(K)
puissance_hanning = np.sum(signal_haning**2)
puissance_hamming = np.sum(signal_hamming**2)
puissance_blackman = np.sum(signal_blackman**2)
puissance_rect = np.sum(rect**2)


plt.figure(figsize=(18, 8))
plt.plot(f,20*np.log10(np.abs(np.fft.fftshift(fft(signal_haning, 10000)))),  label=f"spectre de hanning {puissance_hanning}")
plt.plot(f,20*np.log10(np.abs(np.fft.fftshift(fft(signal_hamming, 10000)))), label=f"spectre de hamming {puissance_hamming}")
plt.plot(f,20*np.log10(np.abs(np.fft.fftshift(fft(signal_blackman, 10000)))),  label=f"spectre de blackman {puissance_blackman}")
plt.plot(f,20*np.log10(np.abs(np.fft.fftshift(fft(rect, 10000)))),  label=f"spectre de rectangle {puissance_hanning}")
plt.ylim([-100, 80])
plt.xlim([-.05, .05])
plt.legend()
plt.show()
```
