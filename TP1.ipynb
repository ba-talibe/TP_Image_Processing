## Partie I : Affichage de matrice. Manipulation d’images.



### Créer et afficher une matrice A de taille 16x16 dans laquelle l’élément aij aura pour valeur i*j.

```python
A = np.zeros((16, 16))
for i in range(A.shape[0]):
    for j in range(A.shape[0]):
        A[i, j] = i*j

A
```

```python

plt.imshow(A,cmap='gray')
plt.colorbar()
```

### Cette question a pour but de s’entraîner 1) au slicing des tableaux numpy et 2) à la représentation d’images couleurs sous la forme de tableaux numpy à trois canaux.

1. Les trois canaux de couleur sont dans l’ordre R, V, B (rouge vert bleu).
2. Le jaune correspond au triplet (R, V, B) = (255, 185, 15)

```python
img = np.zeros((32, 32, 3), dtype=np.uint8)

img[:5, 20:] = [255, 185, 15]
img[5:10, :15] = [255, 0, 0]
img[20:, ::] = [0, 0, 255]

plt.imshow(img,cmap='gray')
plt.colorbar()
```

## Partie III : Profil et projection.

1. Prendre une image NdG de votre choix : tracer un profil vertical et horizontal et commenter. Pour séparer les graphiques, on peut utiliser la fonction plt.figure(k), où k est le numéro de la figure :

```python
img_ndg2 = plt.imread("images/ancien.jpg")

img_ndg2.shape
```

```python
plt.imshow(img_ndg2, cmap="gray")
```

```python
profileH = img_ndg2[56, :]
profileV= img_ndg2[:, 192]
plt.figure(0)
plt.plot(profileH)
plt.figure(1)
plt.plot(profileV)
```

Pour le profile horizontal en 1 : Nous Avons une courbe montagneuse car en arriere plan a haut de l'image nous avon une succesion de plans clairs et de plans foncés 

Pour le  profile Vertical en 2 : un avons une courbe montagneuse mais decroissante car sur le heut de l'image est plus tot claire et au fur et a mesure qu'on descant : ca s'assombrit.

### 2. Calculer et tracer la projection horizontale (somme de tous les profils lignes) et la projection verticale (somme de tous les profils colonnes). On pourra utiliser la fonction np.sum ou cumsum.

```python
sum_v_profile = np.sum(img_ndg2, axis=0)
sum_h_profile = np.sum(img_ndg2, axis=1)

plt.figure(0)
plt.plot(sum_h_profile)
plt.title("sum_h_profile")
plt.figure(1)
plt.plot(sum_v_profile)
plt.title("sum_v_profile")
```
